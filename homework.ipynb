2.1 Bundling boleh memiliki 2 category yang sama<br>
3.Urutan tidak menjadi faktor pembeda (c1 - c2 - c3 = c2 - c1 - c3).<br>
Tentukan ada berapa jumlah kemungkinan bundling yang bisa terbentuk?<br>
Berikut list code category list_id_category = [c1, c2, c3, c4, c5, c6, c7 , c8, c9, c10] (10 poin)

Langkah-langkah:

a. Buatlah variable kosong untuk menampung kemungkinan kombinasi

b. Gunakan penerapan nested for (for dalam for)

c. Buatlah logic pengecekan apakah kombinasi 3 categori sudah ada pada variable inisiasi awal

d. Manfaatkan fungsi sort untuk mengurutkan urutan kombinasi

```python
#No 3
list_id_category=['c1','c2','c3','c4','c5','c6','c7','c8','c9','c10']
#variabel kosong untuk menampung kombinasi
bundling=[]

# cek kombinasi dari 3 kategori cat1,cat2,cat3
for cat1 in list_id_category:
    for cat2 in list_id_category:
        for cat3 in list_id_category:
            if (cat1 !=cat2 !=cat3):
                temp_bundling=sorted([cat1,cat2,cat3]) # mengurutkan kombinasi
                if temp_bundling not in bundling:
                    bundling.append(temp_bundling)

print (len(bundling),' kombinasi')
```

5. Menggunakan dataset rakamin_store_city_monthly! Buatlah 1 kolom baru pada dengan
teknik apply lambda tanpa function def (cukup one liner) dengan nama kolom is_gmail
dengan value True jika menggunakan gmail dan False jika selain gmail.
Tentukan ada berapa user yang menggunakan email gmail pada tahun 2018!
 (14 poin)

Langkah-langkah:<br>
        1. Gunakan teknik apply lambda untuk membuat membuat kolom tahun dengan cara slicing index<br>
        2. Gunakan teknik apply lambda yang fokus di satu kolom untuk membuat kolom baru<br>
        3. Untuk melakukan pengecekan apakah gmail atau tidak bisa menggunakan fungsi in

```python
#No 5
import pandas as pd
df=pd.read_csv('rakamin_store_city_monthly.csv',delimiter=';')

df.head()
```

```python
df['tahun']=df['month'].apply(lambda x: x[:4]) #membuat kolom tahun dengan slicing index

#mengecek apakah menggunakan gmail
df['is_gmail']=df.apply(lambda x:True if 'gmail'in x['email'] else False, axis=1)
df.head()
```

```python
# menentukan user yang menggunakan gmail pada tahun 2018
df[df['month'].str.contains('2018')].groupby('is_gmail').agg({'user_id':'nunique'})
```

6. Menggunakan dataset superstore! Buatlah function yang bisa menghitung Jumlah Sales
dan Proﬁt pada suatu tahun dan region tertentu dengan default tahun 2017. <br>
Tentukan total sales dan proﬁt:<br>
    1. South Region pada tahun 2016<br>
    2. West Region pada tahun 2017

Langkah-langkah:
a. Buatlah function dengan 2 parameter, salah satu parameter nya merupakan default
value 2017
b. Gunakan column Sales dan Proﬁt sebagai metric yang di agregasi (group by)

```python
#No 6
import pandas as pd
df=pd.read_excel('superstore.xls')
df.head()
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Year']=df['Order Date'].dt.year
df.head()
```

```python
def sales_profit (regional, year=2017):
  regional=regional.lower()
  df_sp=df[(df['Region'].str.lower() == regional) & (df['Year'] == year)].reset_index()

  total=df_sp.groupby(['Region','Year']).agg({'Sales':'sum','Profit':'sum'})
  return total
```

kolom baru!

 (20 poin)

```python
import pandas as pd
df=pd.read_excel('superstore.xls')
df.head()
```

```python
df['Year']=df['Order Date'].apply(lambda x:x.year)
df_2017=df[df['Year']==2017]
df_2017=df_2017.groupby(['Customer ID']).agg({'Sales':'sum'}).reset_index()
df_2017.describe()
```

```python
def segmentasi_cust(row):
    if row < 604.910000:
        segment = 'Low'
    elif 604.910000 <= row >= 1334.641600:
        segment = 'Mid'
    else:
        segment = 'High'
    return segment
```

```python
df_2017['segment']=df_2017['Sales'].apply(lambda x:segmentasi_cust(x))
df_2017
```

```python
df_2017.groupby(['segment']).agg({'Customer ID':'count'}).reset_index()
```
